```python
# Check for NaN values in the 'neighborhood' column
neighborhood_nan = data['description'].isna()
nan_count = neighborhood_nan.sum()
print("Number of NaN values:", nan_count)
```

```python
# Removing the stop words

# List of words to remove
words_to_remove = [
    "آسانسور ندارد",
    "بدون آسانسور",
    "پارکینگ ندارد",
    "بدون پارکینگ",
    "انباری ندارد",
    "بدون انباری"
]

# Function to remove specified words from a string
def remove_words(text):
    if isinstance(text, str):
        for word in words_to_remove:
            text = text.replace(word, "")
        return text.strip()
    else:
        return text

# Apply the remove_words function to the 'description' column
data['description'] = data['description'].apply(remove_words)

data
```

```python

# List of words and their corresponding scores
keywords_and_scores = {
    "پارکینگ": 1,
    "انباری": 1,
    "آسانسور": 1,
    "کولر گازی": 1,
    "کولرگازی": 1,
    "دونبش": 1,
    "دو نبش": 1,
    "وان": 1,
    "لابی": 1,
    "نورگیر": 1
}

# Function to calculate the total score for a description
def calculate_score(description):
    score = 0
    for keyword, keyword_score in keywords_and_scores.items():
        if keyword in description:
            score += keyword_score
    return score

# Apply the calculate_score function to the 'description' column and store in 'description_score' column
data['description_score'] = data['description'].apply(calculate_score)

data
```
